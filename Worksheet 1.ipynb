```python
import numpy as np

empty_array = np.empty((2, 2))
print("Empty array 2x2: \n", empty_array)

ones_array = np.ones((4, 2))
print("\n All-one array 4x2: \n", ones_array)

filled_array = np.full((3, 3), fill_value=7)
print("\n Filled array 3x3 with 7: \n", filled_array)

sample_array = np.array([[5, 6, 7], [8, 9, 10]])
zeros_like_array = np.zeros_like(sample_array)
print("\n Zeros like sample_array: \n", zeros_like_array)

ones_like_array = np.ones_like(sample_array)
print("\n Ones like sample_array: \n", ones_like_array)

new_list = [1, 2, 3, 4]
array_from_list = np.array(new_list)
print("\n Array from list: \n ", array_from_list)
```

Task 2

```python
import numpy as np

arr1 = np.arange(10, 50)
print("Array from 10 to 49: \n", arr1)

arr2 = np.arange(9).reshape(3, 3)
print("\n 3x3 matrix 0 to 8: \n", arr2)

identity_matrix = np.eye(3)
print("\n 3x3 Identity matrix: \n", identity_matrix)

random_array = np.random.random(30)
mean_value = random_array.mean()
print("\n Random array of size 30: \n", random_array)
print("Mean of the array:", mean_value)

random_10x10 = np.random.random((10, 10))
min_val = random_10x10.min()
max_val = random_10x10.max()
print("\n 10x10 Random array: \n", random_10x10)
print("Min:", min_val, "Max:", max_val)

zero_array = np.zeros(10)
zero_array[4] = 1
print("\n Zero array with 5th element as 1: \n", zero_array)

arr = np.array([1, 2, 0, 0, 4, 0])
reversed_arr = arr[::-1]
print("\n Reversed array: \n", reversed_arr)

border_array = np.ones((5, 5))
border_array[1:-1, 1:-1] = 0
print("\n 2D array with 1 on border and 0 inside: \n", border_array)

checkerboard = np.zeros((8, 8), dtype=int)
checkerboard[1::2, ::2] = 1
checkerboard[::2, 1::2] = 1
print("\n 8x8 Checkerboard pattern: \n", checkerboard)
```

```python
import numpy as np

x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

add_xy = x + y
print("x + y: \n", add_xy)

sub_xy = x - y
print("\n x - y: \n", sub_xy)

mult_x = x * 3
print("\n x * 3: \n", mult_x)

square_x = x ** 2
print("\n x squared: \n", square_x)

dot_vw = np.dot(v, w)
dot_xv = np.dot(x, v)
dot_xy = np.dot(x, y)
print("\n Dot product v.w:", dot_vw)
print("Dot product x.v: \n", dot_xv)
print("Dot product x.y: \n", dot_xy)

concat_row = np.concatenate((x, y), axis=0)
concat_col = np.concatenate((v.reshape(-1,1), w.reshape(-1,1)), axis=1)
print("\n Concatenate x and y along rows: \n", concat_row)
print("\n Concatenate v and w along columns: \n", concat_col)

try:
    concat_xv = np.concatenate((x, v), axis=0)
except ValueError as e:
    concat_xv = str(e)
print("\n Attempt to concatenate x and v: \n", concat_xv)
```

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

A_inv = np.linalg.inv(A)
identity_check = np.dot(A, A_inv)
print("1. A . A^-1 = \n", identity_check)

print("Is close to identity: \n", np.allclose(identity_check, np.eye(2)))

AB = np.dot(A, B)
BA = np.dot(B, A)
print("\n 2. AB = \n", AB)
print("BA = \n", BA)
print("Is AB equal to BA?", np.array_equal(AB, BA))

AB_T = (AB).T
BT_AT = np.dot(B.T, A.T)
print("\n 3. (AB)^T = \n", AB_T)
print("B^T . A^T = \n", BT_AT)
print("Is (AB)^T equal to B^T A^T?", np.allclose(AB_T, BT_AT))


A_eq = np.array([[2, -3, 1],
                 [1, -1, 2],
                 [3, 1, -1]])

B_eq = np.array([-1, -3, 9])

A_eq_inv = np.linalg.inv(A_eq)
X = np.dot(A_eq_inv, B_eq)
print("\n Solution using inverse method: x, y, z = \n", X)

check = np.dot(A_eq, X)
print("Check A*X = B: \n", check)
```
